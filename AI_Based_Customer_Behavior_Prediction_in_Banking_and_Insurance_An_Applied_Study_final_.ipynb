## Generate Banking Signals

### Subtask:
Synthesize banking-related features for each customer and month, including 'Deposits', 'Withdrawals', 'CardSpend', 'BalanceAvg', 'OverdraftCount', 'DigitalLogins', 'BranchVisits', 'CallCenterContacts', and binary product flags like 'HasLoan', 'HasCreditCard', 'HasSavings'.


**Reasoning**:

````python
new_cols = [
    'Deposits', 'Withdrawals', 'CardSpend', 'BalanceAvg', 'OverdraftCount',
    'DigitalLogins', 'BranchVisits', 'CallCenterContacts', 'HasLoan',
    'HasCreditCard', 'HasSavings'
]
for col in new_cols:
    df_customers[col] = 0

print(f"Initialized new banking signal columns in df_customers. Current shape: {df_customers.shape}")
print(df_customers.head())
````

**Reasoning**:
Now that the banking signal columns are initialized, I will iterate through each customer and then each month to generate realistic banking behavior and populate these columns in the `df_customers` DataFrame, following the instructions for income, deposits, withdrawals, card spend, balance, overdrafts, digital logins, branch visits, call center contacts, and binary product flags.



````python
customer_data = []

for customer_id in df_customers['CustomerID'].unique():
    # Generate baseline income for the customer
    customer_income = np.random.normal(loc=50000, scale=20000)
    customer_income = max(10000, customer_income) # Ensure income is at least 10,000

    # Get customer-level demographic info for this customer
    customer_demographics = df_customers[df_customers['CustomerID'] == customer_id].iloc[0]
    age_band = customer_demographics['AgeBand']
    customer_segment = customer_demographics['CustomerSegment']

    current_balance = 0 # Starting balance for the customer

    # Determine initial product holdings based on demographics
    has_loan_prob = 0.3 # Base probability
    has_credit_card_prob = 0.5
    has_savings_prob = 0.8

    if '18-25' in age_band: # Younger age band, higher credit card prob, lower loan/savings
        has_credit_card_prob *= 1.2
        has_loan_prob *= 0.8
        has_savings_prob *= 0.9
    elif '65+' in age_band: # Older age band, lower credit card prob, higher savings, moderate loan
        has_credit_card_prob *= 0.7
        has_loan_prob *= 0.9
        has_savings_prob *= 1.1

    if 'Premium' in customer_segment: # Premium segment, higher product take-up
        has_credit_card_prob *= 1.1
        has_loan_prob *= 1.1
        has_savings_prob *= 1.1
    elif 'Retail' in customer_segment: # Retail segment, moderate product take-up
        has_credit_card_prob *= 1.0
        has_loan_prob *= 1.0
        has_savings_prob *= 1.0

    initial_has_loan = np.random.rand() < min(1.0, max(0.0, has_loan_prob))
    initial_has_credit_card = np.random.rand() < min(1.0, max(0.0, has_credit_card_prob))
    initial_has_savings = np.random.rand() < min(1.0, max(0.0, has_savings_prob))

    for month_idx in range(N_MONTHS):
        # Deposits (monthly income portion with noise)
        deposits = np.random.normal(customer_income / 12, customer_income / 30)
        deposits = max(0, deposits) # Ensure deposits are non-negative

        # Withdrawals (fraction of deposits)
        withdrawals = deposits * np.random.uniform(0.6, 0.95)
        withdrawals = min(withdrawals, current_balance + deposits + 500) # Prevent excessively large withdrawals if balance is low

        # CardSpend (fraction of deposits, higher for Premium segment)
        card_spend_factor = np.random.uniform(0.1, 0.4)
        if 'Premium' in customer_segment: # Premium customers spend more on cards
            card_spend_factor = np.random.uniform(0.3, 0.6)
        card_spend = deposits * card_spend_factor

        # BalanceAvg (running balance)
        current_balance += deposits - withdrawals - card_spend # Simplified for this example
        current_balance = max(0, current_balance) # Balance cannot go below 0 for simplicity
        balance_avg = current_balance # Using current balance as average for simplicity

        # OverdraftCount (higher if balance is low or negative, for simplicity using a threshold)
        overdraft_count = 0
        if balance_avg < 1000 and np.random.rand() < 0.2: # 20% chance of overdraft if low balance
            overdraft_count = np.random.randint(1, 4)
        elif np.random.rand() < 0.02: # Small chance of overdraft regardless
            overdraft_count = np.random.randint(1, 2)

        # DigitalLogins, BranchVisits, CallCenterContacts
        digital_logins = np.random.randint(0, 16)
        branch_visits = np.random.randint(0, 4)
        call_center_contacts = np.random.randint(0, 3)

        # Product Flags - keep consistent after initial assignment, with small chance of change
        has_loan = initial_has_loan
        has_credit_card = initial_has_credit_card
        has_savings = initial_has_savings

        # Simulate slight changes over time with small probability
        if np.random.rand() < 0.05: # 5% chance of product status changing per month
            has_loan = not has_loan
        if np.random.rand() < 0.05: # 5% chance of product status changing per month
            has_credit_card = not has_credit_card
        if np.random.rand() < 0.05: # 5% chance of product status changing per month
            has_savings = not has_savings

        customer_data.append({
            'CustomerID': customer_id,
            'MonthIndex': month_idx,
            'Deposits': deposits,
            'Withdrawals': withdrawals,
            'CardSpend': card_spend,
            'BalanceAvg': balance_avg,
            'OverdraftCount': overdraft_count,
            'DigitalLogins': digital_logins,
            'BranchVisits': branch_visits,
            'CallCenterContacts': call_center_contacts,
            'HasLoan': int(has_loan),
            'HasCreditCard': int(has_credit_card),
            'HasSavings': int(has_savings)
        })

df_banking_signals = pd.DataFrame(customer_data)

# Merge the generated banking signals back into the main df_customers DataFrame
df_customers = pd.merge(df_customers.drop(columns=new_cols), df_banking_signals, on=['CustomerID', 'MonthIndex'], how='left')

print("Banking signals generated and merged successfully.")
print(f"Updated DataFrame shape: {df_customers.shape}")
print(df_customers.head())
````

````python
np.random.seed(42) # Ensure reproducibility for target generation

# 6. Create binary 'HasClaim' target variable
df_customers['HasClaim'] = ((df_customers['ClaimCount12m'] > 0) | (df_customers['ClaimAmount12m'] > 0)).astype(int)

# 5. Calculate the probability of 'LoanDefault' (only for customers with loans)
default_prob = np.zeros(len(df_customers))
loan_customers_idx = df_customers['HasLoan'] == 1

# Base default probability
default_prob[loan_customers_idx] = 0.05 # Base for customers with loans

# Adjust based on banking behavior
default_prob[loan_customers_idx & (df_customers['OverdraftCount'] > 0)] += 0.08
default_prob[loan_customers_idx & (df_customers['BalanceAvg'] < 500)] += 0.05
default_prob[loan_customers_idx & (df_customers['Withdrawals'] > df_customers['Deposits'] * 1.2)] += 0.07 # Withdrawals significantly higher than deposits

# Cap probability
default_prob = np.clip(default_prob, 0, 0.5) # Max 50% probability

df_customers['LoanDefault'] = (np.random.rand(len(df_customers)) < default_prob).astype(int)

# Only set LoanDefault to 1 if HasLoan is 1 and the random draw is true
df_customers.loc[df_customers['HasLoan'] == 0, 'LoanDefault'] = 0

# 2. Calculate the probability of 'Churn'
churn_prob = np.full(len(df_customers), 0.03) # Base churn probability

# Adjust based on banking behavior
churn_prob[df_customers['OverdraftCount'] > 0] += 0.07
churn_prob[df_customers['BalanceAvg'] < 100] += 0.05
churn_prob[(df_customers['DigitalLogins'] < 2) | (df_customers['DigitalLogins'] > 10)] += 0.03 # Very low or very high engagement

# Adjust based on demographics
churn_prob[df_customers['AgeBand'].isin(['18-25', '26-35'])] += 0.02 # Younger age bands might churn more
churn_prob[df_customers['CustomerSegment'] == 'Retail'] += 0.01

churn_prob = np.clip(churn_prob, 0, 0.4)
df_customers['Churn'] = (np.random.rand(len(df_customers)) < churn_prob).astype(int)

# 3. Calculate the probability of 'PolicyLapse'
lapse_prob = np.full(len(df_customers), 0.04) # Base lapse probability

# Adjust based on insurance behavior
lapse_prob[df_customers['PaymentDelayDays'] > 0] += 0.10
lapse_prob[df_customers['PremiumAmount'] > 300] += 0.03 # Very high premiums might lead to lapse

# Adjust based on demographics
lapse_prob[df_customers['AgeBand'].isin(['18-25'])] += 0.03 # Younger might lapse more
lapse_prob[df_customers['AgeBand'].isin(['65+'])] -= 0.02 # Older might lapse less

lapse_prob = np.clip(lapse_prob, 0, 0.4)
df_customers['PolicyLapse'] = (np.random.rand(len(df_customers)) < lapse_prob).astype(int)

# 4. Determine the probability of 'CrossSellAccept'
cross_sell_prob = np.full(len(df_customers), 0.08) # Base cross-sell probability

# Adjust based on demographics and existing products
cross_sell_prob[df_customers['CustomerSegment'] == 'Premium'] += 0.05
cross_sell_prob[df_customers['AgeBand'].isin(['36-45', '46-55'])] += 0.03 # Middle age might be more open
cross_sell_prob[df_customers['HasLoan'] == 1] += 0.02 # Existing loan indicates trust
cross_sell_prob[df_customers['HasCreditCard'] == 1] += 0.01
cross_sell_prob[df_customers['HasSavings'] == 1] += 0.01

cross_sell_prob = np.clip(cross_sell_prob, 0, 0.3)
df_customers['CrossSellAccept'] = (np.random.rand(len(df_customers)) < cross_sell_prob).astype(int)


print("Target variables generated successfully.")
print("\nClass Imbalance Check:")
for col in target_cols:
    print(f"{col}:\n{df_customers[col].value_counts(normalize=True)}\n")

print(f"Updated DataFrame shape: {df_customers.shape}")
print(df_customers.head())
````

Aggregate banking and insurance features (e.g., deposits, balance, premiums, claims) per customer over the feature window by calculating mean, median, standard deviation, and trend-based features (e.g., slope of spend, logins). **Reasoning**:
The first instruction is to split the `df_customers` DataFrame into `df_features` and `df_outcomes` based on the predefined `FEATURE_WINDOW_MONTHS` and `N_MONTHS`. This will isolate the data for feature generation and label extraction, preventing data leakage.



````python
df_features = df_customers[df_customers['MonthIndex'] < FEATURE_WINDOW_MONTHS].copy()
df_outcomes = df_customers[df_customers['MonthIndex'] >= FEATURE_WINDOW_MONTHS].copy()

print(f"df_customers shape: {df_customers.shape}")
print(f"df_features shape: {df_features.shape}")
print(f"df_outcomes shape: {df_outcomes.shape}")

print("First 5 rows of df_features:")
print(df_features.head())
print("\nFirst 5 rows of df_outcomes:")
print(df_outcomes.head())
````

**Reasoning**:
I need to identify the numerical banking and insurance features, then group `df_features` by 'CustomerID' and calculate the mean, median, and standard deviation for these features, renaming the columns to reflect the aggregations. This will create `df_aggregated_features`.



````python
numerical_features = [
    'Deposits', 'Withdrawals', 'CardSpend', 'BalanceAvg', 'OverdraftCount',
    'DigitalLogins', 'BranchVisits', 'CallCenterContacts', 'PolicyTenureMonths',
    'PremiumAmount', 'PaymentDelayDays', 'ClaimCount12m', 'ClaimAmount12m'
]

# Group by CustomerID and calculate mean, median, and std
df_aggregated_features = df_features.groupby('CustomerID')[numerical_features].agg(['mean', 'median', 'std'])

# Flatten the MultiIndex columns and rename them appropriately
df_aggregated_features.columns = ['_'.join(col).strip() for col in df_aggregated_features.columns.values]
df_aggregated_features = df_aggregated_features.reset_index()

print("Aggregated features (mean, median, std) generated successfully.")
print(f"Shape of df_aggregated_features: {df_aggregated_features.shape}")
print(df_aggregated_features.head())
````

**Reasoning**:
Now, I will calculate trend (slope) features for each numerical feature over the feature window by grouping `df_features` by 'CustomerID' and applying `scipy.stats.linregress` to find the slope of each numerical feature against `MonthIndex`.



````python
from scipy.stats import linregress

def calculate_slope(series):
    if len(series) < 2:
        return np.nan
    # Ensure the index is treated as numerical for regression
    x = series.index.values
    y = series.values
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    return slope

df_trend_features_list = []

for customer_id, group in df_features.groupby('CustomerID'):
    customer_trends = {'CustomerID': customer_id}
    for feature in numerical_features:
        # We need MonthIndex to be the x-variable for the slope calculation
        # So we reset index or ensure MonthIndex is part of the group's index
        temp_df = group[['MonthIndex', feature]].set_index('MonthIndex')
        slope_value = calculate_slope(temp_df[feature])
        customer_trends[f'{feature}_slope'] = slope_value
    df_trend_features_list.append(customer_trends)

df_trend_features = pd.DataFrame(df_trend_features_list)

print("Trend (slope) features generated successfully.")
print(f"Shape of df_trend_features: {df_trend_features.shape}")
print(df_trend_features.head())
````

**Reasoning**:
Now I will extract recency features for each customer by selecting the values of the numerical features from the last month of the feature window, as specified in instruction 5.



````python
df_recency_features = df_features[df_features['MonthIndex'] == FEATURE_WINDOW_MONTHS - 1].copy()
df_recency_features = df_recency_features[['CustomerID'] + numerical_features]
df_recency_features.columns = ['CustomerID'] + [f'{col}_recency' for col in numerical_features]

print("Recency features generated successfully.")
print(f"Shape of df_recency_features: {df_recency_features.shape}")
print(df_recency_features.head())
````

**Reasoning**:
Now I will calculate the 'InteractionIntensity' for each customer by summing 'CallCenterContacts' and 'BranchVisits' over the entire feature window from `df_features` as specified in instruction 6.



````python
df_interaction_intensity = df_features.groupby('CustomerID')[['CallCenterContacts', 'BranchVisits']].sum().reset_index()
df_interaction_intensity['InteractionIntensity'] = df_interaction_intensity['CallCenterContacts'] + df_interaction_intensity['BranchVisits']
df_interaction_intensity = df_interaction_intensity[['CustomerID', 'InteractionIntensity']]

print("Interaction intensity features generated successfully.")
print(f"Shape of df_interaction_intensity: {df_interaction_intensity.shape}")
print(df_interaction_intensity.head())
````

### Justification for Customer-Based Splitting Strategy

The customer-based splitting strategy is crucial for maintaining data integrity and preventing data leakage in scenarios where multiple observations (e.g., monthly records) belong to the same entity (customer).

**Reasons for this strategy:**

1.  **Preventing Data Leakage**: If a customer's data appeared in both the training and testing sets, the model could inadvertently learn patterns specific to that customer from the training data and then perform artificially well on that same customer's data in the test set. This would lead to an overestimation of the model's true generalization performance.

2.  **Realistic Evaluation**: In real-world applications, we often predict outcomes for *new* or *unseen* customers. By ensuring that all data points belonging to a particular customer are confined to a single split (train, validation, or test), we simulate this real-world scenario more accurately. This provides a more robust and realistic evaluation of how the model would perform on completely new customers.

3.  **Temporal Consistency (Implicit)**: While the `MonthIndex` provides a time component, ensuring a customer's entire history (within the feature window) is in one split helps to evaluate the model's ability to generalize to customer profiles rather than just specific time points. Although not a strict time-series split, it respects the individuality of customer behavior patterns.

4.  **Stratification for Representativeness**: Stratifying the splits based on a key target variable (like 'Churn') ensures that the proportion of positive and negative classes for that target is maintained across the training, validation, and testing sets. This is especially important for imbalanced datasets, as it helps prevent a split from having a disproportionately low number of positive cases, which could hinder effective model training and evaluation.

## Conduct Business Impact Simulation

### Subtask:
Define intervention costs and retention values for churn/lapse targets. Simulate the expected net benefit for various top-k targeting strategies (e.g., 1%, 5%, 10% of customers) based on predicted probabilities from the best performing models. **Reasoning**:

These parameters are crucial for the net benefit simulation.



````python
INTERVENTION_COST_CHURN = 50
RETENTION_VALUE_CHURN = 1000

INTERVENTION_COST_LAPSE = 30
RETENTION_VALUE_LAPSE = 700

RETENTION_RATE = 0.3 # Assuming a 30% retention rate for intervened customers

print(f"Intervention Cost for Churn: ${INTERVENTION_COST_CHURN}")
print(f"Retention Value for Churn: ${RETENTION_VALUE_CHURN}")
print(f"Intervention Cost for Lapse: ${INTERVENTION_COST_LAPSE}")
print(f"Retention Value for Lapse: ${RETENTION_VALUE_LAPSE}")
print(f"Assumed Retention Rate for Intervened Customers: {RETENTION_RATE*100}%")
````

**Reasoning**:
Now that the business parameters are defined, I will create the `simulate_net_benefit` function as specified in instruction 3. This function will calculate the net benefit for a given top-k targeting strategy, taking into account predicted probabilities, actual outcomes, intervention costs, retention values, and a retention rate.



````python
def simulate_net_benefit(probabilities, actual_outcomes, cost_per_customer, value_per_retained_customer, k_percentage, retention_rate):
    # Combine probabilities and actual outcomes into a DataFrame for sorting
    simulation_df = pd.DataFrame({
        'probability': probabilities,
        'actual_outcome': actual_outcomes
    })

    # Sort customers by their predicted probabilities (descending order)
    # For churn/lapse, higher probability means higher risk of the negative event, so we target those with highest prob.
    simulation_df = simulation_df.sort_values(by='probability', ascending=False).reset_index(drop=True)

    # Determine the number of customers to target
    num_customers = len(simulation_df)
    num_targeted = int(num_customers * (k_percentage / 100))

    if num_targeted == 0:
        return 0 # No customers targeted, no benefit

    # Select the top k_percentage of customers
    targeted_customers_df = simulation_df.head(num_targeted)

    # Identify customers within the targeted group who would have actually had the negative outcome
    actual_negative_outcomes_in_targeted = targeted_customers_df[targeted_customers_df['actual_outcome'] == 1]
    num_actual_negative_outcomes_targeted = len(actual_negative_outcomes_in_targeted)

    # Calculate expected retained customers
    # Only a fraction (retention_rate) of those who would have had the negative outcome are retained
    expected_retained_customers = num_actual_negative_outcomes_targeted * retention_rate

    # Calculate total intervention cost
    total_intervention_cost = num_targeted * cost_per_customer

    # Calculate gross benefit from retained customers
    gross_benefit = expected_retained_customers * value_per_retained_customer

    # Calculate net benefit
    net_benefit = gross_benefit - total_intervention_cost

    return net_benefit

print("simulate_net_benefit function defined.")
````

**Reasoning**:
Now that the `simulate_net_benefit` function is defined, I will use it to calculate the net benefit for a range of top-k percentages (from 1% to 100% in 1% increments) for both 'Churn' and 'PolicyLapse' targets, storing these results as specified in instruction 4 of the subtask.



````python
k_percentages = np.arange(1, 101, 1) # 1% to 100%

churn_net_benefits = []
policy_lapse_net_benefits = []

# Simulate for Churn
print("Simulating net benefit for Churn...")
for k in k_percentages:
    net_benefit = simulate_net_benefit(
        probabilities=best_probabilities['Churn'],
        actual_outcomes=y_test['Churn'],
        cost_per_customer=INTERVENTION_COST_CHURN,
        value_per_retained_customer=RETENTION_VALUE_CHURN,
        k_percentage=k,
        retention_rate=RETENTION_RATE
    )
    churn_net_benefits.append(net_benefit)

# Simulate for PolicyLapse
print("Simulating net benefit for PolicyLapse...")
for k in k_percentages:
    net_benefit = simulate_net_benefit(
        probabilities=best_probabilities['PolicyLapse'],
        actual_outcomes=y_test['PolicyLapse'],
        cost_per_customer=INTERVENTION_COST_LAPSE,
        value_per_retained_customer=RETENTION_VALUE_LAPSE,
        k_percentage=k,
        retention_rate=RETENTION_RATE
    )
    policy_lapse_net_benefits.append(net_benefit)

print("Net benefit simulations for Churn and PolicyLapse completed.")
````

**Reasoning**:
Now that the net benefits for various top-k percentages have been simulated, I will plot the uplift curves for both 'Churn' and 'PolicyLapse' targets, and identify the `best_k_percentage` that yields the maximum net benefit, as specified in instructions 5 and 6 of the subtask.



````python
import matplotlib.pyplot as plt

# Plotting for Churn
plt.figure(figsize=(10, 6))
plt.plot(k_percentages, churn_net_benefits, marker='o', linestyle='-', color='blue')
plt.title('Uplift Curve for Churn Prediction')
plt.xlabel('Percentage of Customers Targeted (k)')
plt.ylabel('Net Benefit ($)')
plt.grid(True)
plt.xticks(np.arange(0, 101, 10))
plt.tight_layout()
plt.show()

# Identify best k for Churn
max_churn_benefit_idx = np.argmax(churn_net_benefits)
best_k_churn = k_percentages[max_churn_benefit_idx]
max_churn_benefit = churn_net_benefits[max_churn_benefit_idx]
print(f"For Churn, the best k-percentage is {best_k_churn}% with a maximum net benefit of ${max_churn_benefit:.2f}")

# Plotting for PolicyLapse
plt.figure(figsize=(10, 6))
plt.plot(k_percentages, policy_lapse_net_benefits, marker='o', linestyle='-', color='green')
plt.title('Uplift Curve for PolicyLapse Prediction')
plt.xlabel('Percentage of Customers Targeted (k)')
plt.ylabel('Net Benefit ($)')
plt.grid(True)
plt.xticks(np.arange(0, 101, 10))
plt.tight_layout()
plt.show()

# Identify best k for PolicyLapse
max_policy_lapse_benefit_idx = np.argmax(policy_lapse_net_benefits)
best_k_policy_lapse = k_percentages[max_policy_lapse_benefit_idx]
max_policy_lapse_benefit = policy_lapse_net_benefits[max_policy_lapse_benefit_idx]
print(f"For PolicyLapse, the best k-percentage is {best_k_policy_lapse}% with a maximum net benefit of ${max_policy_lapse_benefit:.2f}")

print("Uplift curves generated and best k-percentages identified for Churn and PolicyLapse.")
````

    if len(series) < 2: # Need at least 2 points to calculate a slope
        return np.nan
    x = series.index.values
    y = series.values
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    return slope

````python
import numpy as np
import pandas as pd
import sys
from faker import Faker
from scipy.stats import linregress
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Ensure faker is installed before importing (if not already in environment)
if 'faker' not in sys.modules:
    !pip install -q faker

# --- 1. Re-define Parameters (from 84b1e0b1) ---
N_CUSTOMERS = 10000
N_MONTHS = 24
NUM_RECORDS = N_CUSTOMERS * N_MONTHS
REGIONS = ['North', 'South', 'East', 'West', 'Central']
CUSTOMER_SEGMENTS = ['Retail', 'SME', 'Corporate', 'Premium']
CHANNELS = ['Agent', 'Online', 'Broker']
FEATURE_WINDOW_MONTHS = 18 # from 8470d4c2
OUTCOME_WINDOW_MONTHS = 6

# Set global random seed for reproducibility
np.random.seed(42)

# --- 2. Re-generate Customer Demographics (from 5d8f2881) ---
faker = Faker()
customer_ids_gen = []
age_bands_gen = []
regions_gen = []
customer_segments_gen = []
age_band_choices = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']

for i in range(N_CUSTOMERS):
    customer_ids_gen.append(f'C{i+1:05d}')
    age_bands_gen.append(np.random.choice(age_band_choices))
    regions_gen.append(np.random.choice(REGIONS))
    customer_segments_gen.append(np.random.choice(CUSTOMER_SEGMENTS))

repeated_customer_ids = np.repeat(customer_ids_gen, N_MONTHS)
repeated_age_bands = np.repeat(age_bands_gen, N_MONTHS)
repeated_regions = np.repeat(regions_gen, N_MONTHS)
repeated_customer_segments = np.repeat(customer_segments_gen, N_MONTHS)
month_index = np.tile(np.arange(N_MONTHS), N_CUSTOMERS)

df_customers = pd.DataFrame({
    'CustomerID': repeated_customer_ids,
    'AgeBand': repeated_age_bands,
    'Region': repeated_regions,
    'CustomerSegment': repeated_customer_segments,
    'MonthIndex': month_index
})

# --- 3. Re-generate Banking Signals (from ffde6fa7) ---
customer_data = []

for customer_id in df_customers['CustomerID'].unique():
    customer_income = np.random.normal(loc=50000, scale=20000)
    customer_income = max(10000, customer_income)
    customer_demographics = df_customers[df_customers['CustomerID'] == customer_id].iloc[0]
    age_band = customer_demographics['AgeBand']
    customer_segment = customer_demographics['CustomerSegment']
    current_balance = 0
    has_loan_prob = 0.3
    has_credit_card_prob = 0.5
    has_savings_prob = 0.8

    if '18-25' in age_band:
        has_credit_card_prob *= 1.2
        has_loan_prob *= 0.8
        has_savings_prob *= 0.9
    elif '65+' in age_band:
        has_credit_card_prob *= 0.7
        has_loan_prob *= 0.9
        has_savings_prob *= 1.1

    if 'Premium' in customer_segment:
        has_credit_card_prob *= 1.1
        has_loan_prob *= 1.1
        has_savings_prob *= 1.1
    elif 'Retail' in customer_segment:
        has_credit_card_prob *= 1.0
        has_loan_prob *= 1.0
        has_savings_prob *= 1.0

    initial_has_loan = np.random.rand() < min(1.0, max(0.0, has_loan_prob))
    initial_has_credit_card = np.random.rand() < min(1.0, max(0.0, has_credit_card_prob))
    initial_has_savings = np.random.rand() < min(1.0, max(0.0, has_savings_prob))

    for month_idx in range(N_MONTHS):
        deposits = np.random.normal(customer_income / 12, customer_income / 30)
        deposits = max(0, deposits)
        withdrawals = deposits * np.random.uniform(0.6, 0.95)
        withdrawals = min(withdrawals, current_balance + deposits + 500)
        card_spend_factor = np.random.uniform(0.1, 0.4)
        if 'Premium' in customer_segment:
            card_spend_factor = np.random.uniform(0.3, 0.6)
        card_spend = deposits * card_spend_factor
        current_balance += deposits - withdrawals - card_spend
        current_balance = max(0, current_balance)
        balance_avg = current_balance
        overdraft_count = 0
        if balance_avg < 1000 and np.random.rand() < 0.2:
            overdraft_count = np.random.randint(1, 4)
        elif np.random.rand() < 0.02:
            overdraft_count = np.random.randint(1, 2)
        digital_logins = np.random.randint(0, 16)
        branch_visits = np.random.randint(0, 4)
        call_center_contacts = np.random.randint(0, 3)
        has_loan = initial_has_loan
        has_credit_card = initial_has_credit_card
        has_savings = initial_has_savings
        if np.random.rand() < 0.05:
            has_loan = not has_loan
        if np.random.rand() < 0.05:
            has_credit_card = not has_credit_card
        if np.random.rand() < 0.05:
            has_savings = not has_savings

        customer_data.append({
            'CustomerID': customer_id,
            'MonthIndex': month_idx,
            'Deposits': deposits,
            'Withdrawals': withdrawals,
            'CardSpend': card_spend,
            'BalanceAvg': balance_avg,
            'OverdraftCount': overdraft_count,
            'DigitalLogins': digital_logins,
            'BranchVisits': branch_visits,
            'CallCenterContacts': call_center_contacts,
            'HasLoan': int(has_loan),
            'HasCreditCard': int(has_credit_card),
            'HasSavings': int(has_savings)
        })

df_banking_signals = pd.DataFrame(customer_data)
df_customers = pd.merge(df_customers, df_banking_signals, on=['CustomerID', 'MonthIndex'], how='left')

# --- 4. Re-generate Insurance Signals (from 3ef8660a) ---
customer_insurance_data = []

for customer_id in df_customers['CustomerID'].unique():
    customer_demographics = df_customers[df_customers['CustomerID'] == customer_id].iloc[0]
    age_band = customer_demographics['AgeBand']
    customer_segment = customer_demographics['CustomerSegment']
    initial_channel = np.random.choice(CHANNELS, p=[0.4, 0.4, 0.2])

    for month_idx in range(N_MONTHS):
        policy_tenure_months = month_idx + 1
        premium_base = np.random.normal(loc=150, scale=30)
        if 'Premium' in customer_segment:
            premium_base *= np.random.uniform(1.1, 1.3)
        elif 'Corporate' in customer_segment:
            premium_base *= np.random.uniform(1.0, 1.1)
        if '18-25' in age_band:
            premium_base *= 0.9
        elif '65+' in age_band:
            premium_base *= 1.2
        premium_amount = max(50, round(premium_base + np.random.normal(0, 10), 2))

        payment_delay_days = 0
        if np.random.rand() < 0.05:
            payment_delay_days = np.random.randint(1, 15)

        claim_count_12m = 0
        claim_amount_12m = 0.0
        if np.random.rand() < 0.03:
            claim_count_12m = np.random.randint(1, 3)
            claim_amount_12m = round(np.random.normal(loc=1500, scale=1000), 2)
            claim_amount_12m = max(100, claim_amount_12m)

        current_channel = initial_channel
        if np.random.rand() < 0.02:
            current_channel = np.random.choice(CHANNELS, p=[0.4, 0.4, 0.2])

        customer_insurance_data.append({
            'CustomerID': customer_id,
            'MonthIndex': month_idx,
            'PolicyTenureMonths': policy_tenure_months,
            'PremiumAmount': premium_amount,
            'PaymentDelayDays': payment_delay_days,
            'ClaimCount12m': claim_count_12m,
            'ClaimAmount12m': claim_amount_12m,
            'Channel': current_channel
        })

df_insurance_signals = pd.DataFrame(customer_insurance_data)
df_customers = pd.merge(df_customers, df_insurance_signals, on=['CustomerID', 'MonthIndex'], how='left')

# --- 5. Re-generate Target Variables (from e1a6dc58) ---
target_cols_initial = ['Churn', 'PolicyLapse', 'CrossSellAccept', 'LoanDefault', 'HasClaim']
for col in target_cols_initial:
    df_customers[col] = 0

df_customers['HasClaim'] = ((df_customers['ClaimCount12m'] > 0) | (df_customers['ClaimAmount12m'] > 0)).astype(int)
default_prob = np.zeros(len(df_customers))
loan_customers_idx = df_customers['HasLoan'] == 1
default_prob[loan_customers_idx] = 0.05
default_prob[loan_customers_idx & (df_customers['OverdraftCount'] > 0)] += 0.08
default_prob[loan_customers_idx & (df_customers['BalanceAvg'] < 500)] += 0.05
default_prob[loan_customers_idx & (df_customers['Withdrawals'] > df_customers['Deposits'] * 1.2)] += 0.07
default_prob = np.clip(default_prob, 0, 0.5)
df_customers['LoanDefault'] = (np.random.rand(len(df_customers)) < default_prob).astype(int)
df_customers.loc[df_customers['HasLoan'] == 0, 'LoanDefault'] = 0

churn_prob = np.full(len(df_customers), 0.03)
churn_prob[df_customers['OverdraftCount'] > 0] += 0.07
churn_prob[df_customers['BalanceAvg'] < 100] += 0.05
churn_prob[(df_customers['DigitalLogins'] < 2) | (df_customers['DigitalLogins'] > 10)] += 0.03
churn_prob[df_customers['AgeBand'].isin(['18-25', '26-35'])] += 0.02
churn_prob[df_customers['CustomerSegment'] == 'Retail'] += 0.01
churn_prob = np.clip(churn_prob, 0, 0.4)
df_customers['Churn'] = (np.random.rand(len(df_customers)) < churn_prob).astype(int)

lapse_prob = np.full(len(df_customers), 0.04)
lapse_prob[df_customers['PaymentDelayDays'] > 0] += 0.10
lapse_prob[df_customers['PremiumAmount'] > 300] += 0.03
lapse_prob[df_customers['AgeBand'].isin(['18-25'])] += 0.03
lapse_prob[df_customers['AgeBand'].isin(['65+'])] -= 0.02
lapse_prob = np.clip(lapse_prob, 0, 0.4)
df_customers['PolicyLapse'] = (np.random.rand(len(df_customers)) < lapse_prob).astype(int)

cross_sell_prob = np.full(len(df_customers), 0.08)
cross_sell_prob[df_customers['CustomerSegment'] == 'Premium'] += 0.05
cross_sell_prob[df_customers['AgeBand'].isin(['36-45', '46-55'])] += 0.03
cross_sell_prob[df_customers['HasLoan'] == 1] += 0.02
cross_sell_prob[df_customers['HasCreditCard'] == 1] += 0.01
cross_sell_prob[df_customers['HasSavings'] == 1] += 0.01
cross_sell_prob = np.clip(cross_sell_prob, 0, 0.3)
df_customers['CrossSellAccept'] = (np.random.rand(len(df_customers)) < cross_sell_prob).astype(int)

# --- 6. Split into Features and Outcomes Windows (from 9a6afebb) ---
df_features = df_customers[df_customers['MonthIndex'] < FEATURE_WINDOW_MONTHS].copy()
df_outcomes = df_customers[df_customers['MonthIndex'] >= FEATURE_WINDOW_MONTHS].copy()

# --- 7. Aggregate Features (from dde85eab, bde5a0f1, 86db6273, 36451217, 0cdd7a64, eafcbf6d) ---
numerical_features = [
    'Deposits', 'Withdrawals', 'CardSpend', 'BalanceAvg', 'OverdraftCount',
    'DigitalLogins', 'BranchVisits', 'CallCenterContacts', 'PolicyTenureMonths',
    'PremiumAmount', 'PaymentDelayDays', 'ClaimCount12m', 'ClaimAmount12m'
]

df_aggregated_features = df_features.groupby('CustomerID')[numerical_features].agg(['mean', 'median', 'std'])
df_aggregated_features.columns = ['_'.join(col).strip() for col in df_aggregated_features.columns.values]
df_aggregated_features = df_aggregated_features.reset_index()

def calculate_slope(series):
    if len(series) < 2:
        return np.nan
    x = series.index.values
    y = series.values
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    return slope

df_trend_features_list = []
for customer_id, group in df_features.groupby('CustomerID'):
    customer_trends = {'CustomerID': customer_id}
    for feature in numerical_features:
        temp_df = group[['MonthIndex', feature]].set_index('MonthIndex')
        # Ensure we're calculating slope for numerical features that have more than one unique value
        if temp_df[feature].nunique() > 1:
            slope_value = calculate_slope(temp_df[feature])
        else:
            slope_value = 0.0 # Assign 0 slope if all values are the same
        customer_trends[f'{feature}_slope'] = slope_value
    df_trend_features_list.append(customer_trends)
df_trend_features = pd.DataFrame(df_trend_features_list)

df_recency_features = df_features[df_features['MonthIndex'] == FEATURE_WINDOW_MONTHS - 1].copy()
df_recency_features = df_recency_features[['CustomerID'] + numerical_features]
df_recency_features.columns = ['CustomerID'] + [f'{col}_recency' for col in numerical_features]

df_interaction_intensity = df_features.groupby('CustomerID')[['CallCenterContacts', 'BranchVisits']].sum().reset_index()
df_interaction_intensity['InteractionIntensity'] = df_interaction_intensity['CallCenterContacts'] + df_interaction_intensity['BranchVisits']
df_interaction_intensity = df_interaction_intensity[['CustomerID', 'InteractionIntensity']]

target_cols_for_labels = ['Churn', 'PolicyLapse', 'CrossSellAccept', 'LoanDefault', 'HasClaim']
df_labels = df_outcomes.groupby('CustomerID')[target_cols_for_labels].max().reset_index()

demographic_cols = ['AgeBand', 'Region', 'CustomerSegment']
df_demographics = df_customers.groupby('CustomerID')[demographic_cols].first().reset_index()

# --- 8. Combine All Data (from 5a654725) ---
df_customer_level_data = df_demographics
df_customer_level_data = pd.merge(df_customer_level_data, df_aggregated_features, on='CustomerID', how='left')
df_customer_level_data = pd.merge(df_customer_level_data, df_trend_features, on='CustomerID', how='left')
df_customer_level_data = pd.merge(df_customer_level_data, df_recency_features, on='CustomerID', how='left')
df_customer_level_data = pd.merge(df_customer_level_data, df_interaction_intensity, on='CustomerID', how='left')
df_customer_level_data = pd.merge(df_customer_level_data, df_labels, on='CustomerID', how='left')

# --- 9. Data Cleaning and Preprocessing (from 4155c292, 823564ef, 73b7ae90) ---
numerical_cols_with_nan = df_customer_level_data.select_dtypes(include=np.number).columns[df_customer_level_data.select_dtypes(include=np.number).isnull().any()].tolist()
categorical_cols_with_nan = df_customer_level_data.select_dtypes(include='object').columns[df_customer_level_data.select_dtypes(include='object').isnull().any()].tolist()
for col in numerical_cols_with_nan:
    df_customer_level_data[col] = df_customer_level_data[col].fillna(df_customer_level_data[col].median())
for col in categorical_cols_with_nan:
    df_customer_level_data[col] = df_customer_level_data[col].fillna(df_customer_level_data[col].mode()[0])

categorical_cols = df_customer_level_data.select_dtypes(include='object').columns.tolist()
categorical_cols = [col for col in categorical_cols if col != 'CustomerID']
df_encoded_features = pd.get_dummies(df_customer_level_data[categorical_cols], drop_first=True)
df_customer_level_data = pd.concat([df_customer_level_data.drop(columns=categorical_cols), df_encoded_features], axis=1)

identifier_col = 'CustomerID'
target_cols_final = ['Churn', 'PolicyLapse', 'CrossSellAccept', 'LoanDefault', 'HasClaim']
all_numeric_cols = df_customer_level_data.select_dtypes(include=np.number).columns.tolist()
numerical_features_to_scale = [col for col in all_numeric_cols if col not in target_cols_final]
scaler = StandardScaler()
df_customer_level_data[numerical_features_to_scale] = scaler.fit_transform(df_customer_level_data[numerical_features_to_scale])

# --- 10. Separate X and y (from efcabbc9) ---
X = df_customer_level_data.drop(columns=[identifier_col] + target_cols_final)
y = df_customer_level_data[target_cols_final]

# --- 11. Create combined target and define target_variables (from 54263881) ---
y_train_full = y.copy() # Make copies to avoid SettingWithCopyWarning
y_val_full = y.copy()
y_test_full = y.copy()

y_train_full['Default/ClaimRisk'] = ((y_train_full['LoanDefault'] == 1) | (y_train_full['HasClaim'] == 1)).astype(int)
y_val_full['Default/ClaimRisk'] = ((y_val_full['LoanDefault'] == 1) | (y_val_full['HasClaim'] == 1)).astype(int)
y_test_full['Default/ClaimRisk'] = ((y_test_full['LoanDefault'] == 1) | (y_test_full['HasClaim'] == 1)).astype(int)

target_variables = ['Churn', 'PolicyLapse', 'CrossSellAccept', 'Default/ClaimRisk']

# --- 12. Split data into training, validation, and testing sets (from c3210fc8) ---
customer_ids = df_customer_level_data[identifier_col].unique()
customer_churn_status = df_customer_level_data.drop_duplicates(subset=[identifier_col]).set_index(identifier_col)['Churn']

X_train_ids, X_temp_ids, _, _ = train_test_split(
    customer_ids,
    customer_churn_status[customer_ids],
    test_size=0.2,
    random_state=42,
    stratify=customer_churn_status[customer_ids]
)

X_val_ids, X_test_ids, _, _ = train_test_split(
    X_temp_ids,
    customer_churn_status[X_temp_ids],
    test_size=0.5,
    random_state=42,
    stratify=customer_churn_status[X_temp_ids]
)

train_df = df_customer_level_data[df_customer_level_data[identifier_col].isin(X_train_ids)]
val_df = df_customer_level_data[df_customer_level_data[identifier_col].isin(X_val_ids)]
test_df = df_customer_level_data[df_customer_level_data[identifier_col].isin(X_test_ids)]

X_train = train_df.drop(columns=[identifier_col] + target_cols_final)
y_train = train_df[target_cols_final]

X_val = val_df.drop(columns=[identifier_col] + target_cols_final)
y_val = val_df[target_cols_final]

X_test = test_df.drop(columns=[identifier_col] + target_cols_final)
y_test = test_df[target_cols_final]

# --- 13. Re-define y_train, y_val, y_test with 'Default/ClaimRisk' (from aded034d) ---
y_train['Default/ClaimRisk'] = ((y_train['LoanDefault'] == 1) | (y_train['HasClaim'] == 1)).astype(int)
y_val['Default/ClaimRisk'] = ((y_val['LoanDefault'] == 1) | (y_val['HasClaim'] == 1)).astype(int)
y_test['Default/ClaimRisk'] = ((y_test['LoanDefault'] == 1) | (y_test['HasClaim'] == 1)).astype(int)

# --- 14. Train Models and Collect Metrics (from 09651144) ---
trained_models = {}
test_predictions = {}
test_probabilities = {}
all_metrics = []

rf_param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

xgb_param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2]
}

for target in target_variables:
    print(f"\n==== Processing Target: {target} ====")

    # --- Logistic Regression ---
    print(f"--- Training Logistic Regression for {target} ---")
    lr_model = LogisticRegression(solver='liblinear', random_state=42, max_iter=1000, class_weight='balanced')
    lr_model.fit(X_train, y_train[target])
    trained_models[f'{target}_LogisticRegression'] = lr_model

    y_pred_lr = lr_model.predict(X_test)
    y_pred_proba_lr = lr_model.predict_proba(X_test)[:, 1]
    test_predictions[f'{target}_LogisticRegression'] = y_pred_lr
    test_probabilities[f'{target}_LogisticRegression'] = y_pred_proba_lr

    metrics = {
        'Target': target,
        'Model': 'Logistic Regression',
        'Accuracy': accuracy_score(y_test[target], y_pred_lr),
        'Precision': precision_score(y_test[target], y_pred_lr, zero_division=0),
        'Recall': recall_score(y_test[target], y_pred_lr, zero_division=0),
        'F1-Score': f1_score(y_test[target], y_pred_lr, zero_division=0),
        'ROC-AUC': roc_auc_score(y_test[target], y_pred_proba_lr)
    }
    all_metrics.append(metrics)

    print(f"Accuracy (LR): {metrics['Accuracy']:.4f}")
    print(f"ROC AUC (LR): {metrics['ROC-AUC']:.4f}")

    # --- RandomForestClassifier ---
    print(f"\n--- Training RandomForestClassifier for {target} ---")
    rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
    rf_random_search = RandomizedSearchCV(
        estimator=rf_model,
        param_distributions=rf_param_dist,
        n_iter=5, # Reduced for quicker execution
        cv=2,       # Reduced for quicker execution
        scoring='roc_auc',
        random_state=42,
        n_jobs=-1,
        verbose=0
    )
    rf_random_search.fit(X_train, y_train[target])
    best_rf_model = rf_random_search.best_estimator_
    trained_models[f'{target}_RandomForest'] = best_rf_model
    print(f"Best RandomForest parameters: {rf_random_search.best_params_}")

    y_pred_rf = best_rf_model.predict(X_test)
    y_pred_proba_rf = best_rf_model.predict_proba(X_test)[:, 1]
    test_predictions[f'{target}_RandomForest'] = y_pred_rf
    test_probabilities[f'{target}_RandomForest'] = y_pred_proba_rf

    metrics = {
        'Target': target,
        'Model': 'RandomForest',
        'Accuracy': accuracy_score(y_test[target], y_pred_rf),
        'Precision': precision_score(y_test[target], y_pred_rf, zero_division=0),
        'Recall': recall_score(y_test[target], y_pred_rf, zero_division=0),
        'F1-Score': f1_score(y_test[target], y_pred_rf, zero_division=0),
        'ROC-AUC': roc_auc_score(y_test[target], y_pred_proba_rf)
    }
    all_metrics.append(metrics)

    print(f"Accuracy (RF): {metrics['Accuracy']:.4f}")
    print(f"ROC AUC (RF): {metrics['ROC-AUC']:.4f}")

    # --- XGBoostClassifier ---
    print(f"\n--- Training XGBoostClassifier for {target} ---")
    pos_count = y_train[target].sum()
    neg_count = len(y_train[target]) - pos_count
    scale_pos_weight = neg_count / pos_count if pos_count > 0 else 1

    xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42, scale_pos_weight=scale_pos_weight)
    xgb_random_search = RandomizedSearchCV(
        estimator=xgb_model,
        param_distributions=xgb_param_dist,
        n_iter=5, # Reduced for quicker execution
        cv=2,       # Reduced for quicker execution
        scoring='roc_auc',
        random_state=42,
        n_jobs=-1,
        verbose=0
    )
    xgb_random_search.fit(X_train, y_train[target])
    best_xgb_model = xgb_random_search.best_estimator_
    trained_models[f'{target}_XGBoost'] = best_xgb_model
    print(f"Best XGBoost parameters: {xgb_random_search.best_params_}")

    y_pred_xgb = best_xgb_model.predict(X_test)
    y_pred_proba_xgb = best_xgb_model.predict_proba(X_test)[:, 1]
    test_predictions[f'{target}_XGBoost'] = y_pred_xgb
    test_probabilities[f'{target}_XGBoost'] = y_pred_proba_xgb

    metrics = {
        'Target': target,
        'Model': 'XGBoost',
        'Accuracy': accuracy_score(y_test[target], y_pred_xgb),
        'Precision': precision_score(y_test[target], y_pred_xgb, zero_division=0),
        'Recall': recall_score(y_test[target], y_pred_xgb, zero_division=0),
        'F1-Score': f1_score(y_test[target], y_pred_xgb, zero_division=0),
        'ROC-AUC': roc_auc_score(y_test[target], y_pred_proba_xgb)
    }
    all_metrics.append(metrics)

    print(f"Accuracy (XGB): {metrics['Accuracy']:.4f}")
    print(f"ROC AUC (XGB): {metrics['ROC-AUC']:.4f}")

# Convert all_metrics list into a pandas DataFrame
df_metrics_summary = pd.DataFrame(all_metrics)

# --- 15. Identify and store best models and probabilities for Churn and PolicyLapse (from ac94c7c2) ---
best_models = {}
best_probabilities = {}

for target in ['Churn', 'PolicyLapse']:
    target_metrics = df_metrics_summary[df_metrics_summary['Target'] == target]
    best_model_row = target_metrics.loc[target_metrics['ROC-AUC'].idxmax()]
    best_model_name = best_model_row['Model']

    if best_model_name == 'Logistic Regression':
        model_key = f'{target}_LogisticRegression'
    elif best_model_name == 'RandomForest':
        model_key = f'{target}_RandomForest'
    elif best_model_name == 'XGBoost':
        model_key = f'{target}_XGBoost'
    else:
        model_key = f'{target}_{best_model_name.replace(" ", "")}'

    best_models[target] = trained_models[model_key]
    best_probabilities[target] = test_probabilities[model_key]

    print(f"\nBest model for {target}: {best_model_name} (ROC-AUC: {best_model_row['ROC-AUC']:.4f})")

print("\nAll data generation, feature engineering, preprocessing, and model training steps completed and variables initialized.")
print("df_customer_level_data, X_train, y_train, X_val, y_val, X_test, y_test, df_demographics, trained_models, test_predictions, test_probabilities, df_metrics_summary, best_models, best_probabilities are all available.")
````
